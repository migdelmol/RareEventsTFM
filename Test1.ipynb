1. We load necessary packages.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

2. Then, as a first step, we define our potential, the well-known Müller-Brown potential.

```python
def muellerpot(x,y):
    A = np.array([-200,-100,-170,15])
    mu = np.array([[[1],[0]],[[0],[0.5]],[[-0.5],[1.5]],[[-1],[1]]])
    sigmainv = np.array([[[1,0],[0,10]],[[1,0],[0,10]],[[6.5,-5.5],[-5.5,6.5]],[[0.7,0.3],[0.3,0.7]]])
    VMB = 0
    pointcoords = np.array([[x,y]])
    for i in range(len(A)):
        VMB += A[i]*np.exp(-np.matmul(np.matmul((pointcoords.T-mu[i]).T,sigmainv[i]),pointcoords.T-mu[i]))
    return VMB[0][0]
```

3. After defining our potential, we plot it:

```python
xx, yy = np.meshgrid(np.linspace(-2,1.5,200),np.linspace(-0.5,2.5,200))
zz = np.zeros(xx.shape)

for i in range(xx.shape[0]):
    for j in range(xx.shape[1]):
        zz[i,j] = muellerpot(xx[i,j],yy[i,j])
```

```python
plt.figure(figsize=(8,4),dpi=100)
plt.contourf(xx,yy,zz,)
plt.axis('scaled')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Müller-Brown potential V(x,y)')
plt.show()
```
